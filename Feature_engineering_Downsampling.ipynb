```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
DATA = pd.read_csv("/content/drive/MyDrive/2022/Stock Prediction/Data/VPB_1.csv", header=0, index_col=1, parse_dates=True)
DATA
```

```python
newData_df = DATA[["high", "low", "open", "close", "adjust", "volume"]]
newData_df
```

```python
Week_Data_df = newData_df.resample("W").sum()
Week_Data_df = Week_Data_df[["adjust", "volume"]]
Week_Data_df.rename(columns={"adjust": "weekly_adjust", "volume": "weekly_volume"}, inplace=True)
#Week_Data_df["DayOfWeek"] = Week_Data_df["date"].dt.weekday
#Week_Data_df.loc[Week_Data_df["DayOfWeek"] ==6 , "date"] = Week_Data_df["date"] - timedelta(days=2)
#Week_Data_df.set_index("date", inplace=True)
#Week_Data_df
```

```python
merger_df = pd.merge(newData_df, Week_Data_df, left_index=True, right_index=True, how="outer")
merger_df.head(50)
```

```python
Month_Data_df = newData_df.resample("M").sum()
Month_Data_df = Month_Data_df[["adjust", "volume"]]
Month_Data_df.rename(columns={"adjust": "monthly_adjust", "volume": "monthly_volume"}, inplace=True)
Month_Data_df
```

```python
merger_df = pd.merge(merger_df, Month_Data_df, left_index=True, right_index=True, how="outer")
merger_df.head(50)
```

```python
aaa_df = merger_df.interpolate(method="spline", order=2)
aaa_df.head(50)
```

```python
aaa_df.fillna(aaa_df.mean(), inplace=True)
aaa_df.head(50)
```
